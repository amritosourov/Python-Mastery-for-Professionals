```python
import requests

r = requests.get('https://api.github.com')
print(r.status_code)
print(r.json)
```

2. GET Request with Query Parameters

```python
import requests

params = {'q': 'requests+language:python'}
response = requests.get('https://api.github.com/search/repositories', params=params)
print(response.status_code)
print(response.json())
```

3. GET Request with Headers

```python
import requests

headers = {'User-Agent': 'my-app/0.0.1'}
response = requests.get('https://api.github.com', headers=headers)
print(response.status_code)
print(response.json())
```

4. GET Request with Authentication

```python
import requests
from requests.auth import HTTPBasicAuth

response = requests.get('https://api.github.com/user', auth=HTTPBasicAuth('username', 'password'))
print(response.status_code)
print(response.json())
```

5. GET Request with Timeout

```python
import requests

try:
    response = requests.get('https://api.github.com', timeout=2)  # timeout in seconds
    print(response.status_code)
    print(response.json())
except requests.exceptions.Timeout:
    print('The request timed out')
```

6. Handling Response Content

```python
import requests

response = requests.get('https://api.github.com')
print(response.text)         # Response content as a string
print(response.content)      # Response content in bytes
print(response.json())       # Response content as JSON (if applicable)
```

7. Handling Errors

```python
import requests

response = requests.get('https://api.github.com/nonexistent-endpoint')
if response.status_code == 404:
    print('Not Found.')
elif response.status_code == 200:
    print('Success!')
else:
    print('Something went wrong.')
```
